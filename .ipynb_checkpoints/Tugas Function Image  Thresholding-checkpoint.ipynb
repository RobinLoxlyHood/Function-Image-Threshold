```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt

img = cv.imread('jenderalsoedirman.jpg',0)

plt.imshow(img, cmap='gray')
plt.show()
```

```python
def my_threshold(img, thresh=127, tipe=1):
    if tipe == 1:
        return((img>thresh)*255).astype("uint8")
    elif tipe == 2:
        return((img<thresh)*255).astype("uint8")
    elif tipe == 3:
        return((img>thresh)*thresh).astype("uint8")
    elif tipe == 4:
        return((img>thresh)*img).astype("uint8")
    elif tipe == 5:
        return((img<thresh)*img).astype("uint8")
    else:
        print("Anda Salah memasukan Parameter")
```

## Testing

```python
THRESH_BINARY= my_threshold(img, 150, 1)
cv.imwrite("jenderalsoedirman_THRESH_BINARY.jpg", THRESH_BINARY)
plt.imshow(THRESH_BINARY, cmap='gray')
plt.show()
```

```python
THRESH_BINARY_INV= my_threshold(img, 150, 2)
cv.imwrite("jenderalsoedirman_THRESH_BINARY_INV.jpg", THRESH_BINARY_INV)
plt.imshow(THRESH_BINARY_INV, cmap='gray')
plt.show()
```

```python
THRESH_TRUNC=my_threshold(img, 150, 3)
cv.imwrite("jenderalsoedirman_THRESH_TRUNC.jpg", THRESH_TRUNC)
plt.imshow(THRESH_TRUNC, cmap='gray')
plt.show()
```

```python
THRESH_TOZERO = my_threshold(img, 150, 4)
cv.imwrite("jenderalsoedirman_THRESH_TOZERO.jpg", THRESH_TOZERO)
plt.imshow(THRESH_TOZERO, cmap='gray')
plt.show()
```

```python
THRESH_TOZERO_INV = my_threshold(img, 150, 5)
cv.imwrite("jenderalsoedirman_THRESH_TOZERO_INV.jpg", THRESH_TOZERO_INV)
plt.imshow(THRESH_TOZERO_INV, cmap='gray')
plt.show()
```
